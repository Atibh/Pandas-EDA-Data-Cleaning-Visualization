# Data Cleaning in Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel(r"F:\DATA\Pandas Data Folder\Customer Call List 7.xlsx")
df
```

# removing duplicates

```python
df=df.drop_duplicates()
df
```

# Deleting columns

```python
df=df.drop(columns="Not_Useful_Column")
df
```

# STRIP FUNCTION (strip,lstrip,rstrip)

```python
#lstrip :- removes all unwanted symbols from string on left side
#rstrip :- removes all unwanted symbols from string on right side    # here taking Last_Name column for ex.

# only remove from left and right limitation
```

```python
# df['Last_Name'].str.strip()  # selecting particular column to apply strip function so that it not apply on whole table 
                             # by default it remove white space

#df['Last_Name'].str.lstrip(["/","..."]) #this will not work in dictionary

#df["Last_Name"]=df['Last_Name'].str.lstrip("/")                       # it will save the value to dataframe of particular column
#df["Last_Name"]=df['Last_Name'].str.lstrip("...")
#df["Last_Name"]=df['Last_Name'].str.rstrip("_")
#df 

#df["Last_Name"]=df['Last_Name'].str.strip("123./_")   # removes all unwanted symbols from string in single line
df["Last_Name"]= df["Last_Name"].str.replace("[./_]","" ,regex= True)
df
```

# REPLACE FUNCTION

```python
#df["Phone_Number"].str.replace('[^a-zA-Z0-9]','',regex=True)          # replacing power a to z lower and upper case and number 0 to 9 with blank space 

df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True)
df

```

# Formatting Phone_Number column (using lambda function)

```python
#df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])   # here we got error due to typeerror some are float other is string we have to change datatype

#str(df['Phone_Number'])            # not working we have cnvert data type of each value not whole column
#df['Phone_Number'].apply(lambda x: str(x))            #converting to string
df['Phone_Number']=df['Phone_Number'].astype(str)
df['Phone_Number']=df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10]) 
df
```

# now removing nan--- and Na--

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '',regex=True)
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '',regex=True)
df
```

# have to study fillna and dropna 

```python
df.dropna(subset=['Phone_Number'],inplace = True)  # not workong remove along particular column not in this case
```

# now changing column Paying Customer

```python
df['Paying Customer']=df['Paying Customer'].replace({'Y':'Yes', 'N':'No'})
df['Do_Not_Contact']=df['Do_Not_Contact'].replace({'Y':'Yes', 'N':'No'})
df
```

# Removing N/a and NaN from entire column



```python
df=df.replace('N/a','')
#df=df.replace('NaN','') #only this did not worked instead we can use fillna in both case


df.fillna('')
```

# removing rows where the Do_Not_Contact has yes value 

```python
#df=df[df['Do_Not_Contact']!='Yes'] 
#df                                                               
                                                                 # we can also use 
                                                                 #'df.drop(df[df['Do_Not_Contact'] == 'Y'].index, inplace=True'
# or we can use loop 

for x in df.index :   
    if df.loc[x,'Do_Not_Contact'] == 'Yes':    # for accessing index we have to loc function
        df.drop(x, inplace=True)               # droping column

df
    
```

```python
#df.fillna('')    # only this will not work we have to specify particular column

#df['Do_Not_Contact'].fillna('', inplace=True)       # removing NaN with blank space

df['Do_Not_Contact'].replace('', pd.NA, inplace=True)     # repalcing blank space 


df
```

# removing column with blank phone number

```python
df=df[df['Phone_Number']!=''] 
df  

#another way to drop null value
#df=df.dropna(subset=['Phone_Number'], inplace = True)
#df
```

```python
df=df.reset_index(drop=True)
df
```
